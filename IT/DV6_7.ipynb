```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('/content/MoviesOnStreamingPlatforms_updated.csv')
df.head()
```

# 1

```python
movies_per_year = df.groupby('Year').size().reset_index(name='Counts')

# Vẽ biểu đồ Line
fig = px.line(movies_per_year, x='Year', y='Counts', title='Số lượng Movies qua mỗi năm')
fig.show()
```

# 2

```python
# Tách các thể loại phim (Genres) và đếm số lượng# Tách các thể loại phim (Genres) và đếm số lượng
genres_count = df['Genres'].str.split(',').explode().value_counts().reset_index(name='Counts')

# Vẽ biểu đồ Bar
fig = px.bar(genres_count, x='Genres', y='Counts', title='Số lượng Movies theo thể loại (Genres)', labels={'index':'Genres'})
fig.show()
```

# 3

```python
# Tính trung bình rating theo thể loại
genres_ratings = df.explode('Genres').groupby('Genres')['IMDb'].mean().sort_values(ascending=False).head(5).reset_index()

# Vẽ biểu đồ Bar
fig = px.bar(genres_ratings, x='Genres', y='IMDb', title='Top 5 Thể loại có rating IMDb cao nhất')
fig.show()
```

# 4

```python
# Đếm số lượng phim trên mỗi nền tảng
platforms = ['Netflix', 'Hulu', 'Prime Video', 'Disney+']
platform_counts = df[platforms].sum().reset_index(name='Counts')

# Vẽ biểu đồ Bar
fig = px.bar(platform_counts, x='index', y='Counts', title='Số lượng phim trên các nền tảng khác nhau', labels={'index':'Platforms'})
fig.show()
```

# 5

```python
# Tính IMDb trung bình theo quốc gia
country_ratings = df.groupby('Country')['IMDb'].mean().sort_values(ascending=False).reset_index()

# Vẽ biểu đồ Bar
fig = px.bar(country_ratings, x='Country', y='IMDb', title='Xếp hạng IMDb trung bình ở các quốc gia khác nhau')
fig.show()
```

# 6

```python


# Đếm số lượng phim theo quốc gia
country_count = df['Country'].value_counts().reset_index(name='Movie Count')
country_count.head()
fig = px.choropleth(country_count,
                    locations='Country',

                    color='Movie Count',
                    hover_name='Country',
                    locationmode='country names',
                    title='Số lượng phim trên mỗi quốc gia')
fig.show()
```

# 8

```python
# Đếm số lượng phim theo nhóm tuổi và năm
age_year_count = df.groupby(['Year', 'Age']).size().reset_index(name='Counts')

# Vẽ biểu đồ Line
fig = px.line(age_year_count, x='Year', y='Counts', color='Age', title='Số lượng phim theo nhóm tuổi qua các năm')
fig.show()
```

# 13

```python
# Tính IMDb trung bình theo nhóm tuổi và năm
age_year_rating = df.groupby(['Year', 'Age'])['IMDb'].mean().reset_index()

# Vẽ biểu đồ Line
fig = px.line(age_year_rating, x='Year', y='IMDb', color='Age', title='IMDb trung bình theo nhóm tuổi mỗi năm')
fig.show()
```

# 14

```python
# Đếm số phim theo nền tảng và năm
platform_year_count = df.melt(id_vars='Year', value_vars=['Netflix', 'Hulu', 'Prime Video', 'Disney+'], var_name='Platform', value_name='Available')
platform_year_count = platform_year_count[platform_year_count['Available'] == 1].groupby(['Year', 'Platform']).size().reset_index(name='Counts')

# Vẽ biểu đồ Line
fig = px.line(platform_year_count, x='Year', y='Counts', color='Platform', title='Số lượng phim chiếu trên các nền tảng qua từng năm')
fig.show()
```

# 15

```python
# Đếm số lượng thể loại theo từng nhóm tuổi
genre_age_count = df.explode('Genres').groupby(['Genres', 'Age']).size().reset_index(name='Counts')

# Vẽ biểu đồ Bar
fig = px.bar(genre_age_count, x='Genres', y='Counts', color='Age', title='Thể loại phim theo từng lứa tuổi')
fig.show()
```

# 16

```python
# Lọc phim thuộc thể loại hài, kịch, kinh dị
genres_filter = df[df['Genres'].str.contains('Comedy|Drama|Horror', case=False, na=False)]

# Đếm số lượng phim trên Netflix và Prime Video thuộc các thể loại trên
platform_genre_count = genres_filter.melt(id_vars='Genres', value_vars=['Netflix', 'Prime Video'], var_name='Platform', value_name='Available')
platform_genre_count = platform_genre_count[platform_genre_count['Available'] == 1].groupby(['Platform', 'Genres']).size().reset_index(name='Counts')

# Vẽ biểu đồ Bar
fig = px.bar(platform_genre_count, x='Genres', y='Counts', color='Platform', title='Số lượng phim trên Netflix và Prime theo thể loại hài, kịch, kinh dị')
fig.show()
```

# 19

```python
# Lọc phim từ năm 2015 trở đi và có trên Prime Video
prime_video_movies = df[(df['Year'] >= 2015) & (df['Prime Video'] == 1)]

# Đếm số lượng phim theo năm và nhóm tuổi
prime_video_age_count = prime_video_movies.groupby(['Year', 'Age']).size().reset_index(name='Counts')

# Vẽ biểu đồ Line
fig = px.line(prime_video_age_count, x='Year', y='Counts', color='Age', title='Số phim thuộc Prime Video theo lứa tuổi từ năm 2015 trở về sau')
fig.show()
```

# 20

```python

# Lọc dữ liệu cho các phim từ năm 2015 trở đi
df_2015 = df[df['Year'] >= 2015]

# Tách các đạo diễn và đếm số lượng phim theo mỗi đạo diễn và năm
director_year_count = df_2015.explode('Directors').groupby(['Year', 'Directors']).size().reset_index(name='Counts')

# Vẽ biểu đồ 3D Scatter
fig = px.scatter_3d(director_year_count, x='Year', y='Directors', z='Counts', color='Counts',
                    title="Số phim của mỗi đạo diễn theo năm từ 2015 trở về sau")
fig.show()
```

# 21

```python


# Lọc phim có IMDb > 7
high_rating_movies = df[df['IMDb'] > 7]

# Đếm số lượng phim trên các nền tảng
platforms = ['Netflix', 'Hulu', 'Prime Video', 'Disney+']
platform_counts = high_rating_movies[platforms].sum()

# Tính tỷ lệ phần trăm
platform_percentages = (platform_counts / platform_counts.sum()) * 100

# Vẽ biểu đồ Pie
fig = go.Figure(data=[go.Pie(labels=platform_percentages.index, values=platform_percentages.values,
                             title="Tỉ lệ % số lượng các phim theo các nền tảng có IMDb > 7")])
fig.show()
```

```python


# Gộp nhóm và đếm số lượng phim theo các trường Age, Year, Netflix, Hulu, Prime Video, Disney+
grouped_data = df.groupby(['Age', 'Year', 'Netflix', 'Hulu', 'Prime Video', 'Disney+']).size().reset_index(name='Counts')

# Vẽ biểu đồ Parallel Coordinates
fig = px.parallel_coordinates(grouped_data, dimensions=['Age', 'Year', 'Netflix', 'Hulu', 'Prime Video', 'Disney+'],
                              color='Counts', title="Tọa độ song song của tuổi và các nền tảng")
fig.show()
```

# 24

```python
import plotly.subplots as sp
import plotly.express as px

# Câu 1: Số lượng Movies qua mỗi năm
movies_per_year = df.groupby('Year').size().reset_index(name='Counts')
fig1 = px.line(movies_per_year, x='Year', y='Counts', title='Số lượng Movies qua mỗi năm')

# Câu 2: Số lượng Movies theo thể loại (Genres)
genres_count = df['Genres'].str.split(',').explode().value_counts().reset_index(name='Counts')
fig2 = px.bar(genres_count, x='Genres', y='Counts', title='Số lượng Movies theo thể loại', labels={'index':'Genres'})

# Câu 3: Top 5 thể loại có rating cao nhất
genres_ratings = df.explode('Genres').groupby('Genres')['IMDb'].mean().sort_values(ascending=False).head(5).reset_index()
fig3 = px.bar(genres_ratings, x='Genres', y='IMDb', title='Top 5 Thể loại có rating IMDb cao nhất')

# Câu 4: Số lượng phim trên các nền tảng khác nhau
platforms = df[['Netflix', 'Hulu', 'Prime Video', 'Disney+']].sum().reset_index(name='Counts')
fig4 = px.bar(platforms, x='index', y='Counts', title='Số lượng phim trên các nền tảng khác nhau', labels={'index':'Platforms'})

# Tạo hệ trục tọa độ 2 dòng, 2 cột
fig = sp.make_subplots(rows=2, cols=2, subplot_titles=('Số lượng Movies qua mỗi năm', 'Số lượng Movies theo thể loại',
                                                       'Top 5 Thể loại có rating IMDb cao nhất', 'Số lượng phim trên các nền tảng'))

# Thêm từng biểu đồ vào hệ trục
for trace in fig1.data:
    fig.add_trace(trace, row=1, col=1)

for trace in fig2.data:
    fig.add_trace(trace, row=1, col=2)

for trace in fig3.data:
    fig.add_trace(trace, row=2, col=1)

for trace in fig4.data:
    fig.add_trace(trace, row=2, col=2)

# Hiển thị hệ trục tọa độ
fig.update_layout(height=800, width=1000, title_text="Các biểu đồ câu 1-4 trên cùng hệ trục tọa độ")
fig.show()
```
